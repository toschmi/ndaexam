# How to use Recurrence.py module 

Please save the Recurrence.py module in your working folder 

```python
import matplotlib.pyplot as plt
import numpy as np
from Recurrence import mi,fnn,rp
```

```python
# Preparing the dataset
#this is a test dataset I have, we can use your own dataset


data=np.loadtxt('lorenz.dat')[:,1][::4] 

# data=np.loadtxt('file')
```

```python
plt.plot(data)
```

## Now calculate the Mututal information and embedding dimension:

From Mutual information (__mi__) we can find the $\tau$, from False nearest neighbours (__fnn__) we find the embedding dimension (__m__)



```python
# let's find the delay for this time series

lags,taus=mi(data,20 )
plt.plot(lags,taus)
plt.axvline(np.argmin(taus[:10]),color='r')
print("time delay for the system is:\n")
print("Tau=%d"%np.argmin(taus[:10]))
```

## Now we need to find out the dimension:
we need to use __fnn__ function and need to pass the $\tau$ value which we have found from __mi__

```python
Dims,M=fnn(data,5,10)

plt.plot(Dims,M)
plt.axvline(np.argmin(M[:10]),color='r')
print("Embedding dimension of the system is:\n")
print("m=%d"%np.argmin(M[:10]))
```

## Now we are ready to calculate Recurrence plot:

With this module(Recurrence.py), you no need to do any embedding stuff, it's already there. You need to pass 4 arguments

Recurrence_plot=rp(data,embedding dimension, delay,threshold, threshold_by="mode") 
for threshold part, we need to select it properly. \

Here we have three modes to select the threshold type: \

threshold_by="distance" \
threshold_by="frr" \ it counts certain percentile of the distance matrix



```python
# Lets plot the recurrence plot

Recurrence_plot=rp(data,9,5,0.05,threshold_by='frr')
```

```python
plt.figure(figsize=(6,6),dpi=150)
plt.imshow(Recurrence_plot,cmap='binary',origin='lower')
plt.xlabel("Time")
plt.ylabel("Time")
```
